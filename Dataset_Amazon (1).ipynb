### FAZENDO A CONEXÃO COM O SQLite





```python
import sqlite3

# Conectar ao banco de dados SQLite
connection = sqlite3.connect('amazon.db')
```

```python
import pandas as pd

# Exibir o DataFrame
pd.read_sql_query("select * from sqlite_master where type='table'", connection)
```

### CRIANDO UMA VARIAVEL PARA ARMAZENAR O BANCO DE DADOS

> Estamos armazendando a tabela Reviews em uma Variável já que o nosso Banco de Dados possui uma única tabela.

> **data['time'], unit='s'** converte o tempo para uma versão legível

>**index_col=["Id"]** Para evitar a duplicação da coluna Id

```python
AM = pd.read_sql("select * from Reviews", connection, index_col=["Id"])
AM
```

```python
AM = pd.read_sql("select * from Reviews", connection, index_col=["Id"])

AM['Time'] = pd.to_datetime(AM['Time'], unit='s')

AM
```

# **Dicionário de Dados**

---



| Variável               | Tipo de Dado         | Descrição                                                                   | Tipo           |
|------------------------|----------------------|-----------------------------------------------------------------------------|----------------|
| Id                     | int                  | Índice único para cada entrada no conjunto de dados.                        | numérico       |
| Product Id             | object (string)      | Valor chave de cada produto na Amazon.                                     | categórico     |
| User Id                | object (string)      | ID dos usuários que avaliaram o produto.                                   | categórico     |
| Profile Name           | object (string)      | Nome do avaliador.                                                          | categórico     |
| Helpfulness Numerator  | int                  | Número de usuários para os quais a avaliação foi útil.                     | numérico       |
| Helpfulness Denominator| int                  | Número de usuários para os quais a avaliação não foi útil.                 | numérico       |

| Time                   | datetime             | Timestamp da avaliação.                                                     | data           |
| Summary                | object (string)      | Resumo da avaliação.                                                        | texto          |
